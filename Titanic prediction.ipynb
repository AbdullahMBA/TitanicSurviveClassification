```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 28
        else:
            return 23
    else:
        return Age
    
```

```python
train['Age'] = train[['Age' , 'Pclass']].apply(impute_age , axis = 1)
```
